```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename='sample.csv'
df=pd.read_csv(filename)
```

```python
df.head()
```

```python
df.info()
```

```python
datagraph=df['Month'].values
print(datagraph)
```

```python
volumegraph=df['Volume'].values
print(volumegraph)
```

```python
plt.plot(datagraph,volumegraph)
```

```python
plt.style.available
```

```python
plt.style.use('classic')
plt.plot(datagraph,volumegraph)
```

```python
plt.style.use('seaborn')
plt.plot(datagraph,volumegraph)
```

```python
plt.plot(datagraph,volumegraph, marker='.',markersize=10)
```

```python
plt.plot(datagraph,volumegraph,c='#000000', marker='.',markersize=10)
```

```python
plt.plot(datagraph,volumegraph, marker='.',markersize=10)
plt.xlabel('Month')
plt.ylabel('Volume')
```

```python
plt.plot(datagraph,volumegraph, marker='.',markersize=10)
plt.xlabel('Month',fontsize=15)
plt.ylabel('Volume',fontsize=15)
```

```python
plt.plot(datagraph,volumegraph, marker='.',markersize=10)
plt.xlabel('Month',fontsize=15)
plt.ylabel('Volume',fontsize=15)
plt.xticks(fontsize=15)
```

```python
plt.plot(datagraph,volumegraph, marker='.',markersize=10)
plt.xlabel('Month',fontsize=15)
plt.ylabel('Volume',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
plt.plot(datagraph,volumegraph)
plt.grid(axis='x')
```

```python
plt.plot(datagraph,volumegraph,label='first')
plt.legend(loc='center right')
```

```python
plt.plot(datagraph,volumegraph,label='first')
plt.legend(loc=(1.02,0))
plt.tight_layout()
plt.savefig('graph1.png',dpi=300)
```
